```python
from datasets import load_dataset, Dataset
from transformers import (AutoTokenizer, AutoModelForSeq2SeqLM,
                          DataCollatorForSeq2Seq, Seq2SeqTrainer,
                          Seq2SeqTrainingArguments)
from peft import LoraConfig, get_peft_model
import torch, bitsandbytes as bnb, random, os
```

```python
raw = load_dataset("issai/kazparc", split="train")   # kk / ru / en / tr / domain …

def mk_pairs(batch):
    inputs, targets = [], []
    for kk, ru, en in zip(batch["kk"], batch["ru"], batch["en"]):
        # пропускаем пустые
        kk, ru, en = kk.strip(), ru.strip(), en.strip()
        if kk and ru:
            inputs += [f"translate kaz to ru: {kk}", f"translate ru to kaz: {ru}"]
            targets += [ru, kk]
        if kk and en:
            inputs += [f"translate kaz to en: {kk}", f"translate en to kaz: {en}"]
            targets += [en, kk]
        if ru and en:
            inputs += [f"translate ru to en: {ru}", f"translate en to ru: {en}"]
            targets += [en, ru]
    return {"input": inputs, "target": targets}

pair_ds = raw.map(mk_pairs, batched=True,
                  remove_columns=raw.column_names, num_proc=os.cpu_count())
data = pair_ds.train_test_split(test_size=0.02, seed=42)

print(f"Total training pairs: {len(data['train']):,}")
print(f"Total valid   pairs: {len(data['test']):,}")
```

```python
MODEL_ID = "google/flan-t5-small"
tokenizer = AutoTokenizer.from_pretrained(MODEL_ID, use_fast=False)

base_model = AutoModelForSeq2SeqLM.from_pretrained(
    MODEL_ID,
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_compute_dtype=torch.float16,
    device_map={"": 0},
)

lora_cfg = LoraConfig(
    r=16,
    lora_alpha=32,
    lora_dropout=0.05,
    target_modules=["q","k","v","o"],   # для T5 этого хватает
    bias="none",
    task_type="SEQ_2_SEQ_LM"
)
model = get_peft_model(base_model, lora_cfg)
model.print_trainable_parameters()
```

```python
MAX_LEN = 256
def tok_fn(batch):
    model_in = tokenizer(batch["input"], max_length=MAX_LEN,
                         truncation=True, padding="max_length")
    with tokenizer.as_target_tokenizer():
        labels = tokenizer(batch["target"], max_length=MAX_LEN,
                           truncation=True, padding="max_length")["input_ids"]
    model_in["labels"] = labels
    return model_in

tok_ds = data.map(tok_fn, batched=True,
                  remove_columns=data["train"].column_names,
                  num_proc=os.cpu_count())
```

```python
args = Seq2SeqTrainingArguments(
    output_dir="flan-t5-kazparc",
    per_device_train_batch_size=16,
    gradient_accumulation_steps=2,
    num_train_epochs=3,
    learning_rate=2e-4,
    fp16=True,
    logging_steps=200,
    eval_strategy  ="epoch",
    predict_with_generate=True,
    save_total_limit=2
)

trainer = Seq2SeqTrainer(
    model=model,
    args=args,
    train_dataset=tok_ds["train"],
    eval_dataset=tok_ds["test"],
    data_collator=DataCollatorForSeq2Seq(tokenizer, model)
)
trainer.train()

model.save_pretrained("flan-t5-kaz-rus-en-lora")
tokenizer.save_pretrained("flan-t5-kaz-rus-en-lora")
```

```python
from transformers import pipeline
pipe = pipeline("text2text-generation",
                model="flan-t5-kaz-rus-en-lora",
                tokenizer=tokenizer, device=0)

print(pipe("translate kaz to ru: Мен сені жақсы көремін.")[0]["generated_text"])
```
